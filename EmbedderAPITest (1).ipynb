#### Usage Example

```python
import requests

# API URL
url = "http://127.0.0.1:8000/encode"
```

```python
# Example input: single text
data_single = {
    "text": "Patient reports headache and nausea",
    "normalize": True
}

# Make the POST request for single text
response_single = requests.post(url, json=data_single)
print("Single text embeddings:")
print(response_single.json())
```

```python
# Example input: multiple texts
data_batch = {
    "text": [
        "Patient reports headache and nausea",
        "Shortness of breath and chest pain observed"
    ],
    "normalize": True
}

# Make the POST request for batch
response_batch = requests.post(url, json=data_batch)
print("\nBatch embeddings:")
print(response_batch.json())
```
